***# Question 2. Make a bar graph for manufacturer type.***

```python
Manufacturer_counts = data['Manufacturer'].value_counts()
plt.bar(Manufacturer_counts.index, Manufacturer_counts)
plt.xticks(rotation=90)
plt.xlabel('Manufacturer')
plt.ylabel('Count')
plt.title('Manufacturer Type Distribution')
plt.figure(figsize=(10, 6))
plt.show()
```

***# Question 3. Make a pie chart for showing vehicle type by percentage.***

```python
vehicle_type_counts = data['Vehicle_type'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(vehicle_type_counts, labels=vehicle_type_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Vehicle Type Distribution')
plt.show()
```

***# Question 4. Calculate average Price of all cars by using correct measure of central Tendency.***

***# Question 5. Calculate the variance and standard deviation for numerical columns such as Sales_in_thousands, __year_resale_value, Price_in_thousands, Engine_size, Horsepower.***

```python
numerical_columns = ['Sales_in_thousands', '__year_resale_value', 'Price_in_thousands', 'Engine_size', 'Horsepower']
variances = data[numerical_columns].var()
std_devs = data[numerical_columns].std()

print("Variance:")
print(variances)

print("\nStandard Deviation:")
print(std_devs)
```

***# Question 6.For the same columns Calculate IQR and identify outliers in these columns
Q1 = df[column].quantile(0.25)
Q3 = df[column].quantile(0.75)
IQR = Q3 - Q1
outliers = df[(df[column] &lt; Q1 - 1.5 * IQR) | (df[column] &gt; Q3 + 1.5 * IQR)]***

```python
for column in numerical_columns:
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1

    outliers = data[(data[column] < Q1 - 1.5 * IQR) | (data[column] > Q3 + 1.5 * IQR)]

    print(f"\nOutliers for {column}:")
    print(outliers)
```

***# Question 7. Make Histograms for all Numerical columns and identify the distribution of type by calculating
skewness;
Note : If the skewness is 0, it indicates a perfectly symmetrical distribution.
If the skewness is negative, it means the distribution is skewed to the left (long left tail).
If the skewness is positive, it means the distribution is skewed to the right (long right tail).***

```python
import matplotlib.pyplot as plt
from scipy.stats import skew

for column in numerical_columns:
    plt.figure(figsize=(8, 6))
    plt.hist(data[column], edgecolor='black')
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()

    skewness = skew(data[column])
    print(f"Skewness for {column}: {skewness}")

    if skewness == 0:
        print("The distribution is perfectly symmetrical.")
    elif skewness < 0:
        print("The distribution is skewed to the left (long left tail).")
    else:
        print("The distribution is skewed to the right (long right tail).")

```
